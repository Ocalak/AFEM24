```python
import json
from sklearn.metrics import mean_absolute_error
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from datetime import datetime
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
from tensorflow.keras.layers import Dense, Dropout,BatchNormalization
from tensorflow.keras.regularizers import l2
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers, Model, Input,regularizers
import statistics
import tensorflow_probability as tfp

os.listdir('/content/')
file_path = '/content/pythondf.csv'
dfload = pd.read_csv(file_path)

dfload = dfload.drop(dfload.columns[0], axis=1)
# Define a function to add '00:00:00' if time component is missing
def add_missing_time_component(date_str):
    if len(date_str) == 10:  # Assuming 'YYYY-MM-DD' format
        return date_str + ' 00:00:00'
    return date_str

# Apply the function to the DateTime column
dfload['DateTime'] = dfload['DateTime'].apply(add_missing_time_component)

# Now convert the DateTime column to datetime format
dfload['DateTime'] = pd.to_datetime(dfload['DateTime'], format='%Y-%m-%d %H:%M:%S', errors='coerce')


# Convert the datetime column to datetime objects
dfload['DateTime'] = pd.to_datetime(dfload['DateTime'])

# Extract hour of the day, day of the week, and month of the year
dfload['hour'] = dfload['DateTime'].dt.hour
dfload['day_of_week'] = dfload['DateTime'].dt.dayofweek  # Monday=0, Sunday=6
dfload['month'] = dfload['DateTime'].dt.month

# Convert these variables to dummy variables
df_hour_dummies = pd.get_dummies(dfload['hour'], prefix='hour').astype(int)
#df_day_dummies = pd.get_dummies(dfload['day_of_week'], prefix='day_of_week').astype(int)
#df_month_dummies = pd.get_dummies(dfload['month'], prefix='month').astype(int)

for lag in [192,216,336,504]:
    dfload[f'load_lag_{lag}'] = dfload['Load_DA'].shift(lag)

for lag in range(1, 4):
    dfload[f'temp_lag_{lag}'] = dfload['temp_actual'].shift(lag)
for lag in [48]:
     dfload[f'temp_lag_{lag}'] =  dfload['temp_actual'].shift(lag)
dfload['temp_min'] =  dfload['temp_actual'].rolling(window=24).min()
dfload['temp_max'] =  dfload['temp_actual'].rolling(window=24).max()

f_combined = pd.concat([dfload, df_hour_dummies], axis=1)
dfx = pd.concat([f_combined.drop(columns=['hour','dwpt_actual','pres_actual','wspd_actual'])], axis=1)







# Load the best hyperparameters
best_params =  {'use_temp_actual': True, 'use_day_of_week': True, 'use_month': True, 'use_load_lag_192': True, 'use_load_lag_216': True, 'use_load_lag_336': True, 'use_load_lag_504': True, 'use_temp_lag_1': True, 'use_temp_lag_2': False, 'use_temp_lag_3': True, 'use_temp_mean_3': False, 'use_temp_mean_8': False, 'use_temp_mean_24': False, 'use_temp_lag_24': False, 'use_temp_lag_48': True, 'use_temp_min': True, 'use_temp_max': True, 'use_hour_0': True, 'use_hour_1': True, 'use_hour_2': True, 'use_hour_3': False, 'use_hour_4': False, 'use_hour_5': False, 'use_hour_6': True, 'use_hour_7': False, 'use_hour_8': True, 'use_hour_9': True, 'use_hour_10': False, 'use_hour_11': False, 'use_hour_12': True, 'use_hour_13': False, 'use_hour_14': True, 'use_hour_15': True, 'use_hour_16': False, 'use_hour_17': True, 'use_hour_18': True, 'use_hour_19': False, 'use_hour_20': False, 'use_hour_21': True, 'use_hour_22': True, 'use_hour_23': True, 'regularize_h1_activation': False, 'regularize_h1_kernel': False, 'neurons_1': 212, 'activation_1': 'relu', 'regularize_h2_activation': False, 'regularize_h2_kernel': False, 'neurons_2': 238, 'activation_2': 'softplus', 'learning_rate': 0.006009436831141169}


window = slice(8785,43825)
train_val_set = dfx[window]
test_set = dfx[43825:]
features = train_val_set.drop(columns=['DateTime',	'Load_DA',	'Load_Act'])

selected_features = [feature for feature, use in best_params.items() if use and feature.startswith('use_')]
selected_features = [feature.replace('use_', '') for feature in selected_features]

# Prepare the feature set based on selected features
X_selected = features[selected_features].values

# Normalize the selected features
scaler = MinMaxScaler()
X_selected_scaled = scaler.fit_transform(X_selected)
target = train_val_set['Load_DA'].values.reshape(-1, 1)
y_reshaped = np.array([target[i:i+192] for i in range(len(target) - 192)])

# Adjust the input features accordingly
X_selected_scaled = X_selected_scaled[:-192]

# Split the data into training and testing sets
X_train, X_val, y_train, y_val = train_test_split(X_selected_scaled, y_reshaped, test_size=0.25, random_state=42)


def create_final_model(input_dim, best_params):
    inputs = Input(shape=(input_dim,))

    # First hidden layer
    hidden_layer1 = layers.Dense(best_params['neurons_1'],
                                 activation=best_params['activation_1'])(inputs)

    # Second hidden layer
    hidden_layer2 = layers.Dense(best_params['neurons_2'],
                                 activation=best_params['activation_2'])(hidden_layer1)

    # Output layer
    outputs = layers.Dense(192, activation='linear')(hidden_layer2)

    # Create the model
    model = Model(inputs=inputs, outputs=outputs)

    # Compile the model
    learning_rate = best_params['learning_rate']
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_absolute_error')

    return model

# Create the model using the best parameters
final_model = create_final_model(X_train.shape[1], best_params)

# Train the model
callbacks = [tf.keras.callbacks.EarlyStopping(patience=50, restore_best_weights=True)]
final_model.fit(X_train, y_train, epochs=1000, callbacks=callbacks, batch_size=128, verbose=1, validation_data=(X_val, y_val))


```

```python
# Make predictions on the test set
features_t = test_set.drop(columns=['DateTime',	'Load_DA',	'Load_Act'])

X_t = features_t[selected_features].values
scaler = MinMaxScaler()
X_s = scaler.fit_transform(X_t)
X_se = X_s[:-192]
X_test= X_se
y_pred = final_model.predict(X_test)
y_target = test_set['Load_DA'].values.reshape(-1, 1)
y_re = np.array([y_target[i:i+192] for i in range(len(y_target) - 192)])
y_rex = y_re[1]
# Evaluate the model's performance
#mae = mean_absolute_error(y_rex.reshape(-1), y_pred.reshape(-1))
print(f"Mean Absolute Error (MAE) on the test set: {mae}")
```
